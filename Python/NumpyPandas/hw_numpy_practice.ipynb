5. В данном массиве замените все числа, меньшие 0 на 0, а большие 255 на 255. Напишите минимум два варианта решения

```python
electron = np.random.randint(-50, 50, 5)
```

```python
nucleus = np.random.randint(200, 300, 5)
```

```python
protium = np.concatenate([electron, nucleus])
protium
```

```python
protium.clip(0, 255)
```

```python
deuterium = np.hstack((electron, nucleus))
deuterium
```

```python
deuterium[deuterium < 0] = 0
deuterium[deuterium > 255] = 255
deuterium
```

```python
tritium = np.vstack((electron, nucleus))
tritium
```

```python
tritium = np.where(tritium < 0, 0, np.where(tritium > 255, 255, tritium))
tritium
```

8. Удалите все np.nan из массива. 
*Для того, чтобы в тестовом массиве появился nan используйте np.nan или деление 0. на 0.*

```python
alps = np.random.randint(-100, 100, 20)
alps
```

```python
ural = np.where(alps < 15, np.nan, alps)
ural
```

```python
# Поиск пропусков
nan_search = np.isnan(ural)
```

```python
# Фильтрация пропущенных значений
himalayas = ural[~nan_search]
himalayas
```

9. Загрузите изображение, сохраненное в формате .npy lena.npy или любое другое изображение в формате png.
Используйте известные вам способы модификации numpy массива для того, чтобы изменить изображение. Отсматривайте результаты, используя функцию для отрисовки изображений plt.imshow().

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("lena.png")
```

```python
img = np.array(img)
```

```python
np.save("lena.npy", img)
```

```python
img = np.load("lena.npy")
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
# транспонирируем первые две оси 
turn = np.transpose(img, (1, 0, 2))
```

```python
plt.imshow(turn)
```

```python
# вырежем глаза
eyes = img[240:290, 225:375]
```

```python
plt.imshow(eyes)
```

```python
# функция subplots дает возможность вывести сразу несколько изображений
# в качестве параметров передаем количество строк и столбцов сетки изображений (1 x 3)
# и размер графиков
fig, ax = plt.subplots(1, 3, figsize = (12, 4))
 
# и выводим каждое изображение по отдельности
 
# сначала только красные оттенки, слой [0]
# также нужно принудительно задать cmap = 'Reds'
ax[0].imshow(img[:,:,0], cmap = 'Reds')
# также зададим заголовок
ax[0].set_title('Оттенки красного')
 
# потом только зеленые, слой [1]
ax[1].imshow(img[:,:,1], cmap = 'Greens')
ax[1].set_title('Оттенки зеленого')
 
# и наконец только синие, слой [2]
ax[2].imshow(img[:,:,2], cmap = 'Blues')
ax[2].set_title('Оттенки синего');
```

```python
# уменьшить изображение в два раза
less_twice = img.reshape(img.shape[0] // 2, 2, img.shape[1] // 2, 2, -1)
less_twice.shape
```

```python
plt.imshow(less_twice[:, 0, :, 0])
```

```python
# объединим по ширине две одинаковые копии уменьшенного в два раза изображения
in_width = np.hstack((less_twice[:, 0, :, 0], less_twice[:, 0, :, 1]))
plt.imshow(in_width)
```

```python
# объединим по высоте две одинаковые копии уменьшенного в два раза изображения
in_height = np.vstack((less_twice[:, 0, :, 0], less_twice[:, 1, :, 0]))
plt.imshow(in_height)
```

```python
# закрасить область
paint_over = img.copy()
lavender = 230, 230, 250
darkslateblue = 72, 61, 139 

start, end = np.arange(100), np.arange(400, 512)

paint = np.concatenate([start, end])

paint_over[paint, :, :] = darkslateblue
paint_over[:, paint, :] = lavender

paint_over[:start[-1], end[0]:, :] = darkslateblue
paint_over[end[0]:, :start[-1] + 1, :] = darkslateblue


plt.imshow(paint_over)
```

```python
bw = Image.open("bw.jpg")
bw = np.array(bw)
np.save("bw", bw)
bw = np.load("bw.npy")
plt.imshow(bw)
```

```python
# если значение пикселя ниже 87, сделаем его черным, выше - белым
binary = np.where( bw < 87, 0, 255)
 
# теперь сравним исходное изображение и изображение после преобразования
fig, ax = plt.subplots(1, 2, figsize = (8, 4))
 
# выведем первое изображение и зададим заголовок
ax[0].imshow(bw,  cmap = 'gray')
ax[0].set_title('До')
 
# для ч/б изображения не забудем про параметр cmap = 'gray'
ax[1].imshow(binary,  cmap = 'gray')
ax[1].set_title('После')
 
plt.show()
```
